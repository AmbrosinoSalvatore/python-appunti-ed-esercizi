```python
import pandas as pd
import matplotlib.pyplot as plt
```

crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
data = {'last_name':['boyer', 'copeland', 'daniel', 'serrano','terrel','vargas'], 'orders_number':[185, 50, 196, 58,24,28]}     
df = pd.DataFrame(data)
df
```

visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:

```python
data = list(df['orders_number'])
names = list(df['last_name'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names,data)
```

colori diversi


```python
data = list(df['orders_number'])
names = list(df['last_name'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names,data,color = 'r')
ax.set_xlabel('last_name')
ax.set_ylabel('orders_number')
fig.suptitle('istogramma')
plt.show()
```

Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
df = df.sort_values('orders_number',ascending = False)
fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'],df['orders_number'])
ax.set_xlabel('last_name')
ax.set_ylabel('orders_number')
fig.suptitle('istogramma')
plt.show()
```

visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
data = {'last_name':['copeland','serrano', 'boyer', 'daniel', 'vargas','terrel'],
    'perc':[9.242144,10.720887,34.195933,36.22905,5.175601,4.436229]}
df = pd.DataFrame(data)
df
```

visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
perc = df['perc']
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=[10,10])
ax = plt.axes()
ax.pie(perc, labels= df['last_name'],autopct='%1.2f%%')
plt.show
```

visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
color = ['c','k','c','m','g','r']
perc = df['perc']
plt.rcParams.update({'font.size':14})
fig = plt.figure(figsize=[10,10])
ax = plt.axes()
ax.pie(perc, labels= df['last_name'],colors=color,autopct='%1.2f%%')
fig.suptitle('grafico a torta')
plt.show
```

venduto (‘quantity’) e quanto ha venduto in termini di $ (‘total_amount’) ogni membro dello staff

visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
fig,(ax,ax2)=plt.subplots(1,2)
fig.set_size_inches(15,8)
ax.bar(df['name'],df['quantity'])
ax2.bar(df['name'],df['total'])
plt.show()
```
